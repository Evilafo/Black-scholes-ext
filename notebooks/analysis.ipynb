```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../src')
from black_scholes import black_scholes_call, black_scholes_put
from binomial_tree import binomial_tree_option
```

```python
S = 100
K = 100
r = 0.05
sigma = 0.2
T = 1

S_values = np.linspace(50, 150, 100)
call_prices = [black_scholes_call(S_val, K, r, sigma, T) for S_val in S_values]
put_prices = [black_scholes_put(S_val, K, r, sigma, T) for S_val in S_values]

plt.figure(figsize=(10, 6))
plt.plot(S_values, call_prices, label="Call Price", color="blue")
plt.plot(S_values, put_prices, label="Put Price", color="red")
plt.axvline(K, linestyle="--", color="black", label="Strike Price")
plt.title("Prix des options en fonction du prix de l'actif sous-jacent")
plt.xlabel("Prix de l'actif sous-jacent (S)")
plt.ylabel("Prix de l'option")
plt.legend()
plt.grid()
plt.show()
```

```python
N = 100
binomial_call_price = binomial_tree_option(S, K, r, sigma, T, N, option_type="call")
binomial_put_price = binomial_tree_option(S, K, r, sigma, T, N, option_type="put")

bs_call_price = black_scholes_call(S, K, r, sigma, T)
bs_put_price = black_scholes_put(S, K, r, sigma, T)

print(f"Black-Scholes Call: {bs_call_price:.2f}, Binomial Call: {binomial_call_price:.2f}")
print(f"Black-Scholes Put: {bs_put_price:.2f}, Binomial Put: {binomial_put_price:.2f}")
```
